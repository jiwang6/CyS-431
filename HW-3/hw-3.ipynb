```python
def prime_factors(n):
    i = 2
    factors = []
    while i ** 2 <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

def list_fermat_pseudo():
    n = int(input("nerd"))
    a = 2

    for i in range(1,n+1):
        # (a ^ b) % p
        if pow(a, i-1, i) == 1:
            factors = prime_factors(i)
            if len(factors) == 1: 
                print(f"{i} Fermat Pseudoprime")
            else:
                print(f"{i} ***** NOT PRIME but Fermat Pseudoprime base {a} with factors {factors}]*****")
        else:
            print(i)

list_fermat_pseudo()
```
